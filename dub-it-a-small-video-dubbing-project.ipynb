# **EXTRACT TEXT FROM THE AUDIO FILE USING WHISPER** #

```python
model = whisper.load_model('large')
```

```python
%%time
txt_imran_urdu = model.transcribe('/kaggle/working/imran_urdu.wav')
txt_imran_urdu=txt_imran_urdu["text"]
print(txt_imran_urdu)

#%%time
txt_news_urdu = model.transcribe('/kaggle/working/news_urdu.wav')
txt_news_urdu=txt_news_urdu["text"]
print(txt_news_urdu)

#%%time
txt_messi_spanish = model.transcribe('/kaggle/working/messi_spanish.wav')
txt_messi_spanish=txt_messi_spanish["text"]
print(txt_messi_spanish)

#%%time
txt_actor1_urdu = model.transcribe('/kaggle/working/actor1_urdu.wav')
txt_actor1_urdu=txt_actor1_urdu["text"]
print(txt_actor1_urdu)

#%%time
txt_actor2_urdu = model.transcribe('/kaggle/working/actor2_urdu.wav')
txt_actor2_urdu=txt_actor2_urdu["text"]
print(txt_actor2_urdu)
```

# **CONVERT ORIGINAL TEXT TO ANOTHER LANGUAGE USING TRANSLATOR** #

```python
translator = Translator()
imran_english = translator.translate(text=txt_imran_urdu,dest="english")
news_english = translator.translate(text=txt_news_urdu,dest="english")
messi_english = translator.translate(text=txt_messi_spanish,dest="english")
actor1_english = translator.translate(text=txt_actor1_urdu,dest="english")
actor2_english = translator.translate(text=txt_actor2_urdu,dest="english")

print(imran_english.text,"\n")
print(news_english.text,"\n")
print(messi_english.text,"\n")
print(actor1_english.text,"\n")
print(actor2_english.text,"\n")
```

```python
data=pd.DataFrame([imran_english.text,news_english.text,messi_english.text,actor1_english.text,actor2_english.text])

print(data.head())

data.to_csv("text_transcriptions.csv",index=False)
```
